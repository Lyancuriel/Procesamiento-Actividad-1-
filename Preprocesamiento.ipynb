```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('cuentas_credicel (1).xlsx')
df
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data1=df.copy()
```

```python
#Primer criterio
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data1=data1.fillna(round(df.median(),1))
data1
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Segundo criterio
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia atras  "backwards fill" ("bfill")
#Filtro por columnas
data1["empresa"] =data1["empresa"].fillna(method="bfill")
data1["fecha_ultimo_pago"] =data1["fecha_ultimo_pago"].fillna(method="bfill")
data1["fecha_pago_proximo"] =data1["fecha_pago_proximo"].fillna(method="bfill")
data1["riesgo"] =data1["riesgo"].fillna(method="bfill")
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Tercer criterio
#Sustituir valores nulos por valores no nulos hacia delante  "forward fill" ("ffill")
#Filtro por columnas
data1["status_cuenta"] =data1["status_cuenta"].fillna(method="ffill")
data1["razones_buro"] =data1["razones_buro"].fillna(method="ffill")
data1["semana_actual"] =data1["semana_actual"].fillna(method="ffill")
data1["codigo_postal"] =data1["codigo_postal"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
data1.info()
```

```python
cuantitativas=data1.iloc[ : , [2,7,8,9,10,11,12,13,16,17,19,20,21,25,27,28,29]]  #Columnas no consecutivas
cuantitativas
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data2= cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data2.copy()
data_clean=data_clean.fillna(round(data2.median(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data_clean.isnull().sum()
valores_nulos
```
